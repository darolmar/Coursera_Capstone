# Segmenting and Clustering Neighborhoods in Toronto

Wikipedia page, https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M, in order to obtain the data that is in the table of postal codes and to transform the data into a pandas dataframe

# Import packages

```python
from bs4 import BeautifulSoup
import re
import requests
import bs4
import pandas as pd
origin = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup = BeautifulSoup(origin, 'lxml')
```

# Get the table

```python
table = soup.find('table', class_ = "wikitable sortable")
df = pd.read_html(str(table))[0]
list = df.drop([0])
list.head()
```

```python
# Ignore cells with a borough that is Not assigned
list1 = list[-list.Borough.isin(['Not assigned'])]
list1.head()
```

```python
# Replace the Neighbourhood that is not assigned with Borough
import numpy as np
list1['Neighborhood'][list1['Neighborhood'] == 'Not assigned'] = list1['Borough']
list1.head()
```

```python
group = list1.groupby(['Postal Code', 'Borough'], as_index = False).agg({'Neighborhood': ['sum']})
```
