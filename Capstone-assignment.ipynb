# Segmenting and Clustering Neighborhoods in Toronto

Wikipedia page, https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M, in order to obtain the data that is in the table of postal codes and to transform the data into a pandas dataframe

# Import packages

```python
from bs4 import BeautifulSoup
import re
import requests
import bs4
import pandas as pd
origin = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup = BeautifulSoup(origin, 'html.parser')
```

# Get the table

```python
table = soup.find('table',{'class':'wikitable sortable'})
# define the dataframe rows and columns
rows = table.find_all('tr')
columns = [v.text.replace('\n','') for v in rows[0].find_all('th')] 
# instantiate the dataframe
df = pd.DataFrame(columns=columns)
for i in range(1,len(rows)):
    tds = rows[i].find_all('td')
    values = [td.text.replace('\n','') for td in tds]
    df=df.append(pd.Series(values,index=columns),ignore_index=True)
```

```python
# Preprocess data
df = df[df.Borough != 'Not assigned']
# Replace the unassigned neighbourhood with its borough
df['Neighborhood'].loc[(df['Neighborhood']=="Not assigned")]=df['Borough']
df
```

```python
group=df.groupby(['Postal Code','Borough']).agg(lambda col:','.join(col))
group = group.reset_index()
group
```
